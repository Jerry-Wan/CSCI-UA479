```python
dirData1 = pd.read_json("http://linserv1.cims.nyu.edu:10000/films?_page=1")
dirData2 = pd.read_json("http://linserv1.cims.nyu.edu:10000/films?_page=2")
dirData3 = pd.read_json("http://linserv1.cims.nyu.edu:10000/films?_page=3")
frames = [dirData1,dirData2,dirData3]
dirData = pd.concat(frames)
```

```python
count = dirData.groupby(['director']).count()['title']
```

```python
score = dirData.groupby(['director']).mean()['rt_score']
```

```python
pd.DataFrame([count,score]).transpose()
```
